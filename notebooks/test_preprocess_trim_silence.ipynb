```python
%matplotlib inline

from glob import glob
import os
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf

list = glob("../dump/train/wavs/*-wave.npy")
list.sort()

files = []

for file_path in list:
    file_name = os.path.basename(file_path)
    file_id = file_name.replace("-wave.npy", "")
    file_folder = file_id[0]
    
    original_path = f"../datasets/{file_folder}/{file_id}.wav"
    files.append((file_id, file_path, original_path))

for file_id, file_path, original_path in files[30:40]:
    clipped = np.load(file_path)
    original, _ = sf.read(original_path)
    
    print(file_id)

    plt.subplot(2, 1, 1)
    plt.plot(original)
    plt.title("original")
    plt.subplot(2, 1, 2)
    plt.plot(clipped)
    plt.title("clippped")
    plt.tight_layout()
    plt.show()
    plt.close()
```
