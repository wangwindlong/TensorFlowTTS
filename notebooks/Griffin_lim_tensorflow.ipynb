```python
from tensorflow_tts.utils import TFGriffinLim

import numpy as np
import matplotlib.pyplot as plt
import yaml
```

```python
with open("../preprocess/ljspeech_preprocess.yaml") as f:
    config = yaml.load(f, Loader=yaml.Loader)
```

```python
griffin_lim = TFGriffinLim(config=config, name="griffin_lim")
```

```python
griffin_lim.setup_stats(np.load("../dump/stats.npy"))
griffin_lim._n_iters = 60
```

```python
norm_mel_spectrogram = np.load("../dump/valid/norm-feats/LJ001-0009-norm-feats.npy")
```

```python
inv_wav = griffin_lim(norm_mel_spectrogram)
```

```python
plt.plot(inv_wav)
```
